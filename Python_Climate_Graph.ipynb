```python
#temperature scatter plot

plt.rcParams["figure.figsize"] = (12,6)

plt.scatter(temperature['Date'],temperature['Value'],color='orange')
plt.title('United States Average Temperatures - EU',fontsize=14)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Temperatures (°F)', fontsize=14)
plt.text(1962.5,52.55, "1950-2020 Mean: 52.4°F", ha='center')
plt.text(1962.5, 53.75, "2000-2020 Mean: 53.5°F", ha='center')
plt.axhline(y=53.5, color= 'b', linestyle = ':')
plt.axhline(y=52.4, color= 'b', linestyle = ':')
plt.show()
```
